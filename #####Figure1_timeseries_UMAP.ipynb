```r
library(Signac)
library(Seurat)
library(EnsDb.Mmusculus.v79)
library(GenomeInfoDb)
library(ggplot2)
library(patchwork)
set.seed(1234)
```

```r
skin <- readRDS("./RDS/230126.E13_PD4_integrated.2-10.reference.motif.chromvar.rds")
```

```r
skin
```

```r
Idents(object = skin) <- "bulk"
```

```r
options(repr.plot.width = 12, repr.plot.height = 9)
```

```r
DimPlot(
      skin,
      reduction = "umap", 
      pt.size = 0.3, sizes.highlight = 1,
      cells.highlight = CellsByIdentities(skin, idents = 'E13.5'),  cols.highlight = "#e06666") + NoLegend() +NoAxes()

ggsave("E13.svg", width =4, height =3.5, units = 'in')
```
